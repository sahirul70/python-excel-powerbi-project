# Exploratory Data Analysis

plotting a bar chart for Industry and it's count

```python
ax = sns.countplot(x = 'Industry',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

plotting a bar chart for gender vs total amount

```python
Revenue_gen = df.groupby(['Industry'], as_index = False)['Revenue'].sum().sort_values(by='Revenue', ascending= True)
```

```python
sns.barplot(x='Industry',y='Revenue', data=Revenue_gen)
```

Note : From above graphs we can see that most of the Revenue & total sales from Bevarage, food, Chemical and Other 

plotting a bar chart for Company Size and it's count

```python
#df['Company Size'].unique().sum().values_count()
%matplotlib inline

ax = sns.countplot(x = 'Company Size',data = df)
sns.set(rc={'figure.figsize':(10,6)})


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Revenue_company = df.groupby(['Company Size'], as_index= False)['Revenue'].sum().sort_values(by='Revenue', ascending=True)

sns.barplot(x='Company Size', y='Revenue',data=Revenue_company )
```

```python
Revenue_by_country = df.groupby(['Country'],as_index= False)['Revenue'].sum()#.reset_index() #.sort_values(by='Revenue', ascending=True)
Revenue_by_country
```

```python
fig = plt.figure(figsize =(10, 7))
plt.title('Country by Revanue with %')
plt.pie(Revenue_by_country['Revenue'], labels =Revenue_by_country['Country'],autopct='%.0f%%')
plt.show()
```
